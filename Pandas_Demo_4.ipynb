```python
country_respondents = df ['Country'].value_counts()
country_respondents
```

```python
country_uses_py = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_py
```

```python
python_df = pd.concat([country_respondents, country_uses_py], axis = 'columns', sort= False)
python_df
```

```python
python_df.rename(columns = {'count': 'Number of Respondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace = True)
python_df
```

```python
python_df['pct_knowsPython'] = (python_df["NumKnowsPython"]/python_df['Number of Respondents'])*100
python_df.sort_values(by = 'pct_knowsPython', ascending = False, inplace = True)
python_df.head(25)
```
